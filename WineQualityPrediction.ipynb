## Importing The Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

## Data Collection

```python
# Loading the dataset to a Pandar DataFrame
df = pd.read_csv('winequality-red.csv')

# First 5 rows of the dataset To check its features
df.head()
```

```python
# No of rows and columns in the dataset
df.shape
```

```python
# Checking for missing values
df.isnull().sum()
```

## Data Analysis and Visualization

```python
# Statistical measures of the dataset
df.describe()
```

```python
# Number of values for each quality
sns.catplot(x='quality' ,data = df , kind='count')
df['quality'].value_counts()
```

```python
# Volatile acidity vs Quality
plot = plt.figure(figsize=(5,5))
sns.barplot(x='quality', y='volatile acidity', data=df)
```

```python
# Citric acid vs Quality
plot = plt.figure(figsize=(5,5))
sns.barplot(x='quality', y='citric acid', data=df)
```

## Data Preprocessing

```python
# Seperating Data and Label

x = df.drop(['quality'],axis=1)
print(x)
```

#### Label Binarization

```python
y = df['quality'].apply(lambda y_value:1 if y_value>=7 else 0)
print(y)
```

```python
y.value_counts()
```

## Train Test Split

```python
x_train , x_test , y_train, y_test = train_test_split(x,y,test_size=0.2 , random_state= 42)
```

```python
print(x.shape , x_train.shape , x_test.shape)
```

```python
print(y.shape , y_train.shape , y_test.shape)
```

## Random Forest Classifier

```python
model = RandomForestClassifier()
model
```

```python
model.fit(x_train,y_train)
```

## Model Evaluation & Accuracy Score

```python
x_test_prediction = model.predict(x_test)
test_data_score = accuracy_score(x_test_prediction , y_test)
print('Accuracy : ', test_data_score)
```

## Building a Predictive System

```python
## Making a predictive System
input_data=(7.3,0.65,0.0,1.2,0.065,15.0,21.0,0.9946,3.39,0.47,10.0)

#changing the data to numpy array
input_data_as_numpy_array = np.asarray(input_data)

#reshape the np array as the prediction is for single instance
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

prediction = model.predict(input_data_reshaped)
print(prediction)

if (prediction[0] == 1):
    print('Good Quality')
else:
    print('Bad Quality')
```
